# Merging all the data into one file

```python
# https://stackoverflow.com/questions/70294391/merge-a-list-of-dataframes-by-one-column-with-reduce-function
merge_configs = [
    {'data': data_weather, 'how': 'inner', 'on': ['season', 'round', 'circuit_id'], 'drop_cols': ['lat', 'long', 'country']},
    {'data': data_results, 'how': 'inner', 'on': ['season', 'round', 'circuit_id', 'url'], 'drop_cols': ['url', 'points', 'status', 'time']},
    {'data': data_driver_standings, 'how': 'left', 'on': ['season', 'round', 'driver'], 'drop_cols': None},
    {'data': data_constructor_standings, 'how': 'left', 'on': ['season', 'round', 'constructor'], 'drop_cols': None},
    {'data': data_qualifying, 'how': 'inner', 'on': ['season', 'round', 'grid'], 'drop_cols': ['driver_name', 'car']}
]

merged_finaldata = reduce(lambda left, right: pd.merge(left, right['data'], how=right['how'], on=right['on']).drop(right['drop_cols'], axis=1) if right['drop_cols'] else pd.merge(left, right['data'], how=right['how'], on=right['on']), merge_configs, data_races)

merged_finaldata.head()
```

# Finding null values in the merged file and cleaning it

```python
merged_finaldata.isna().sum()/len(merged_finaldata)
```

```python
for col in ['driver_points', 'driver_wins', 'driver_standings_pos', 'constructor_points', 'constructor_wins' , 'constructor_standings_pos']:
    merged_finaldata[col].fillna(0, inplace = True)
    merged_finaldata[col] = merged_finaldata[col].map(lambda x: int(x))
```

# Calculating Drivers Age and cleaning the qualifying time

```python
# https://stackoverflow.com/questions/48231758/merge-pandas-dataframes-based-on-date
merged_finaldata['date'] = pd.to_datetime(merged_finaldata.date) 
merged_finaldata['date_of_birth'] = pd.to_datetime(merged_finaldata.date_of_birth) 
merged_finaldata['driver_age'] = merged_finaldata.apply(lambda x: relativedelta(x['date'], x['date_of_birth']).years, axis=1) 
merged_finaldata.drop(['date', 'date_of_birth'], axis = 1, inplace = True) 
```

```python
merged_finaldata['qualifying_time'] = merged_finaldata['qualifying_time'].apply(lambda x: float(x.split(':')[1]) if (len(str(x)) >= 8) else x)
merged_finaldata['qualifying_time'] = merged_finaldata['qualifying_time'].astype(float)
```

# Calculating Delta in qualifiying time

```python
# https://stackoverflow.com/questions/66843142/use-fillna-and-lambda-function-in-pandas-to-replace-nan-values
final_dumy = (
    merged_finaldata
    .drop(
        merged_finaldata[(merged_finaldata['circuit_id'] == 'bahrain') & (merged_finaldata['season'] == 2020)].index
    )
    .assign(
        qualifying_time=lambda x: x['qualifying_time'].fillna(
            x.groupby(['circuit_id', 'constructor', 'season'])['qualifying_time'].transform('mean')
        ).fillna(
            x.groupby(['circuit_id', 'constructor'])['qualifying_time'].transform('mean')
        ).fillna(
            x.groupby(['circuit_id'])['qualifying_time'].transform('mean')
        )
    )
    .loc[lambda x: x['qualifying_time'] != 0]
    .sort_values(['season', 'round', 'grid'])
    .assign(
        qualifying_time_diff=lambda x: x.groupby(['season', 'round'])['qualifying_time'].diff(),
        qualifying_time=lambda x: x.groupby(['season', 'round'])['qualifying_time_diff'].cumsum().fillna(0)
    )
    .drop('qualifying_time_diff', axis=1)
)
```

# Removing dummy columns with not enough values

```python
df_dum = pd.get_dummies(final_dumy, columns = ['circuit_id', 'nationality', 'constructor'] )

for col in df_dum.columns:
    if 'nationality' in col and df_dum[col].sum() < 140:
        df_dum.drop(col, axis = 1, inplace = True)
        
    elif 'constructor' in col and df_dum[col].sum() < 140:
        df_dum.drop(col, axis = 1, inplace = True)
        
    elif 'circuit_id' in col and df_dum[col].sum() < 70:
        df_dum.drop(col, axis = 1, inplace = True)
    
    else:
        pass
```

```python
final_dumy.to_csv('data/dummies_removed.csv', index=False)
```
